```python
import os
import shapefile
shp1 = shapefile.Reader('/home/karthik/Western_ghats/shapefiles/wg_boundary_modifiedNBR.shp')
shapes = shp1.shapes();
region = ee.Geometry.Polygon(shapes[0].points);
```

```python
m.addLayer(wg_shp, {}, 'ID')
print(shapes[0].points)
```

```python
##LST##
##LST day rasters##
lstdval  =  ee.ImageCollection('MODIS/006/MOD11A2').select(['LST_Day_1km']).filterDate('2000-01-01','2001-01-01')
lstdval1 =  lstdval.filterBounds(region)

lstdvalList = lstdval1.toList(lstdval1.size())
lstdvalSize = lstdvalList.getInfo()

##LST day Quality layer##
lstdvalqc  =  ee.ImageCollection('MODIS/006/MOD11A2').select(['QC_Day']).filterDate('2000-01-01','2001-01-01')
lstdvalqc1 =  lstdvalqc.filterBounds(region)

lstdvalqcList = lstdvalqc1.toList(lstdvalqc1.size())
lstdvalqcSize = lstdvalqcList.getInfo()
```

```python
##LST night rasters##
lstnval  =  ee.ImageCollection('MODIS/006/MOD11A2').select(['LST_Night_1km']).filterDate('2000-01-01','2001-01-01')
lstnval1 =  lstnval.filterBounds(region)

lstnvalList = lstnval1.toList(lstnval1.size())
lstnvalSize = lstnvalList.getInfo()

##LST night Quality layer##
lstnvalqc  =  ee.ImageCollection('MODIS/006/MOD11A2').select(['QC_Night']).filterDate('2000-01-01','2001-01-01')
lstnvalqc1 =  lstnvalqc.filterBounds(region)

lstnvalqcList = lstnvalqc1.toList(lstdvalqc1.size())
lstnvalqcSize = lstnvalqcList.getInfo()
```

```python
print(len(lstnvalSize))
print(lstnvalqcList.get(0).getInfo())
```

```python
for i in range(0, len(lstnvalqcSize)):
    
    #Day#
    lstd_img = ee.Image(lstdvalList.get(i)).float()
    filn = 'LST_day_'+ ee.Date(ee.Image(lstdvalList.get(i)).get('system:time_start')).format('YYYY-MM-dd').getInfo()+'.tif'
    filename = os.path.join('/home/karthik/Western_ghats/',filn)
    print("LST_day:", i)
    geemap.ee_export_image(lstd_img, filename = filename, scale = 1000, region = region, file_per_band = True, crs = 'EPSG:4326')
    
    
    lstd_qc = ee.Image(lstdvalqcList.get(i)).float()
    filn1 = 'QC_day_'+ ee.Date(ee.Image(lstdvalqcList.get(i)).get('system:time_start')).format('YYYY-MM-dd').getInfo()+'.tif'
    filename1 = os.path.join('/home/karthik/Western_ghats/',filn1)
    print("QC_day:", i)
    geemap.ee_export_image(lstd_qc, filename = filename1, scale = 1000, region = region, file_per_band = True, crs = 'EPSG:4326')
    
    ##Night##
    lstn_img = ee.Image(lstnvalList.get(i)).float()
    fil = 'LST_night_'+ ee.Date(ee.Image(lstnvalList.get(i)).get('system:time_start')).format('YYYY-MM-dd').getInfo()+'.tif'
    filname = os.path.join('/home/karthik/Western_ghats/',fil)
    print("LST_night:", i)
    geemap.ee_export_image(lstn_img, filename = filname, scale = 1000, region = region, file_per_band = True, crs = 'EPSG:4326')
    
    
    lstn_qc = ee.Image(lstnvalqcList.get(i)).float()
    fil1 = 'QC_night_'+ ee.Date(ee.Image(lstnvalqcList.get(i)).get('system:time_start')).format('YYYY-MM-dd').getInfo()+'.tif'
    filname1 = os.path.join('/home/karthik/Western_ghats/',fil1)
    print("QC_night:", i)
    geemap.ee_export_image(lstn_qc, filename = filname1, scale = 1000, region = region, file_per_band = True, crs = 'EPSG:4326')
```
