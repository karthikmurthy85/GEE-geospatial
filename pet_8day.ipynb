```python
import os
import shapefile
shp1 = shapefile.Reader('/home/karthik/Western_ghats/shapefiles/wg_boundary_modifiedNBR.shp')
shapes = shp1.shapes();
region = ee.Geometry.Polygon(shapes[0].points);
```

```python
m.addLayer(wg_shp, {}, 'ID')
```

```python
pet = ee.ImageCollection('MODIS/006/MOD16A2').select(['PET']).filterDate('2001-01-01','2020-10-01')
petList = pet.toList(pet.size())
petSize = petList.getInfo()
print(len(petSize))

petQC = ee.ImageCollection('MODIS/006/MOD16A2').select(['ET_QC']).filterDate('2001-01-01','2020-10-01')
petQCList = petQC.toList(petQC.size())
petQCSize = petQCList.get(0).getInfo()
print(petQCSize)
```

```python
for i in range(0, len(petSize)):
    
    ##VEgeation area (total biomass area)
    pet_img = ee.Image(petList.get(i)).float()
    filn = 'PET_'+ ee.Date(ee.Image(petList.get(i)).get('system:time_start')).format('YYYY-MM-dd').getInfo()+'.tif'
    filename = os.path.join('/home/karthik/Western_ghats/PET/',filn)
    print("PET:", i)
    geemap.ee_export_image(pet_img, filename = filename, scale = 500, region = region, file_per_band = True, crs = 'EPSG:4326')
    
    ##Quality layer
    petQC_img = ee.Image(petQCList.get(i)).float()
    filn1 = 'QA_'+ ee.Date(ee.Image(petQCList.get(i)).get('system:time_start')).format('YYYY-MM-dd').getInfo()+'.tif'
    filename1 = os.path.join('/home/karthik/Western_ghats/PET/',filn1)
    print("Pixel Reliability:", i)
    geemap.ee_export_image(petQC_img, filename = filename1, scale = 500, region = region, file_per_band = True, crs = 'EPSG:4326')
 
```
